### 1. AM wave

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.0, 5.0, 0.01)
y1 = np.cos(np.pi*t); #message
y2 = np.sin(10*np.pi*t); #carrier
m=0.8; # modulation index
ym=(1+m*y1)*y2; #Modulated wave

plt.figure(1)
plt.subplot(311)
plt.plot(t,y1, 'r--')
plt.xlabel('t')
plt.ylabel('m(t)')
plt.title('Message')
plt.grid(True, which='both')

plt.subplot(312)
plt.plot(t,y2, 'r--')
plt.xlabel('t')
plt.ylabel('yc(t)')
plt.title('Carrier')
plt.grid(True, which='both')

plt.subplot(313)
plt.plot(t,ym, 'r--')
plt.xlabel('t')
plt.ylabel('ym(t)')
plt.title('Modulated Wave')
plt.grid(True, which='both')
plt.show()
```

Observation:
- The first subplot shows the message signal, a cosine wave representing the modulating signal.  
- The second subplot displays the carrier signal, a high-frequency sine wave.  
- The third subplot illustrates the amplitude-modulated (AM) wave, where the carrier amplitude varies according to the message signal.  
- The modulation index (`m = 0.8`) controls the extent of amplitude variation in the modulated signal.  
- Grid lines are enabled for better visualization of signal variations.  


### 2. Spectrum of AM wave

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.0, 5.0, 0.01)
y1 = np.cos(np.pi*t); #message
y2 = np.sin(10*np.pi*t); #carrier
m=0.8; # modulation index
ym=(1+m*y1)*y2; #Modulated wave

f = np.arange(0.0, 5.0, 0.01)
Z=np.fft.fft(ym);

Z=abs(Z);

plt.plot(Z[1:100], 'r--')
plt.show()
```

 Observation:
- The FFT plot represents the frequency components of the amplitude-modulated (AM) signal.  
- Sidebands appear due to modulation, indicating frequency content beyond the carrier.  
- The modulation index (`m = 0.8`) affects the strength of the sidebands.  
- The carrier frequency and sidebands are visible in the frequency domain representation.

- The FM signal appears as a frequency-modulated version of the original wave.  


### Spectrum of FM wave

```python
import numpy as np
import matplotlib.pyplot as plt
#Generate t = np.arange(0, (np.pi), 0.0001)
fc = 50
b = 8
data = np.cos(np.pi * 1 * t)

phi = fc*t + b * data
fm = np.cos(phi)

f = np.arange(0, (np.pi), 0.001)
Z=np.fft.fft(fm);

Z=abs(Z);
plt.plot(Z[1:200], 'r--')
plt.show()
```

 Observation:
- The FFT plot displays the frequency components of the FM signal.
- Multiple peaks indicate the sidebands caused by frequency modulation.
- The carrier frequency and modulating signal influence the spectral spread.
- The red dashed plot represents the magnitude spectrum of the FM signal.
